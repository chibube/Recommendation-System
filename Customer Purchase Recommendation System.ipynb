```python
# importing necessary libraries
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#loading the excel dataset
df = pd.read_excel('Online Retail.xlsx')
df.head()
```

```python
#exploring the dataset
df.info()
```

```python
print('Number of columns: ', df.shape[1])
print('Number of rows: ', df.shape[0])
print('Number of null values: ', df.isnull().sum().values.sum())
print('Dataframe features: \n', df.columns.tolist())
print('Dataframe unique values: \n', df.nunique())
```

```python
#creating a Milestone for our dataset before we start cleaning it
df2 = df.copy()
```

```python
df2.head()
```

```python
#deleting NaN rows
df2 = df2.dropna(axis = 0)
df.head()
```

```python
df2.isnull().sum()
```

```python
#deleting negative quantity and price
df2 = df2[df2['Quantity'] > 0]
df2 = df2[df2['UnitPrice'] > 0]
df2.describe()
```

```python
# creating a user to product matrix using pivot tables
User_product_matrix = pd.pivot_table(df2, values = 'Quantity', index = ['CustomerID'], columns = ['StockCode'], aggfunc = 'sum')
User_product_matrix
```

```python
#converting the relationships to binary as we are only interested in knowning that a user bought a product
customer_item_matrix = User_product_matrix.applymap(lambda x: 1 if x > 0 else 0)
customer_item_matrix.head()                                    
```

### Creating the Cosine Similarities


```python
#computing pairwise cosine similarities of users - this compares the similarities between 2 customers
user_user_sim_matrix = pd.DataFrame(cosine_similarity(customer_item_matrix))
user_user_sim_matrix
```

```python
#reinstating the customer IDs to easily identify users with similarities
user_user_sim_matrix.columns = customer_item_matrix.index
user_user_sim_matrix.columns
```

```python
#reinstating the customer IDs to easily identify users with similarities
user_user_sim_matrix.index = customer_item_matrix.index
user_user_sim_matrix.head()
```

```python
#showing the top 10 most similar customer-customer product relationship
def ten_sim_customers(x):
        top_ten_customers = lambda x: user_user_sim_matrix[x].sort_values(ascending = False)
        top_ten_customers = pd.DataFrame(top_ten_customers(x)).iloc[1:11]
        return top_ten_customers
    
ten_sim_customers(12348.0)
```

```python
#getting the recommendations for each user
sim_cus_products = set(User_product_matrix.columns[User_product_matrix.loc[13027] > 0])
customers_purchases = set(User_product_matrix.columns[User_product_matrix.loc[12348] > 0])
recommendation = sim_cus_products - customers_purchases
recommendation
```

```python
#retrieving the items to recommend to the user based on the most similar customer
def rec_item_desc(x): 
    return df2.loc[df2['StockCode'] == x, 'Description'].unique()
    
rec_item_desc(21212)
items_to_recommend = list(map(rec_item_desc, recommendation))
pd.DataFrame(items_to_recommend)
```

```python
#retrieving the items to recommend to the user based on the most similar customer with a neater code
df2.loc[df2['StockCode'].isin(recommendation), ['StockCode', 'Description']].drop_duplicates().set_index('StockCode')
```
